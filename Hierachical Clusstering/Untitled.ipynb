```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Customers.csv')
```

```python
x = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Eucledian Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity = 'euclidean',linkage = 'ward')
```

```python
y_hc =hc.fit_predict(x)
```

```python
plt.scatter(x[y_hc ==0,0],x[y_hc==0,1],s=100,color='red',label='Cluster1')
plt.scatter(x[y_hc ==1,0],x[y_hc==1,1],s=100,color='magenta',label='Cluster2')
plt.scatter(x[y_hc ==2,0],x[y_hc==2,1],s=100,color='pink',label='Cluster3')
plt.scatter(x[y_hc ==3,0],x[y_hc==3,1],s=100,color='green',label='Cluster4')
plt.scatter(x[y_hc ==4,0],x[y_hc==4,1],s=100,color='blue',label='Cluster5')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
